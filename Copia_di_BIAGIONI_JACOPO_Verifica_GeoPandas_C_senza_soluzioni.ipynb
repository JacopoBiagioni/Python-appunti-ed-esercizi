```
# caricare il geodataset dei quartieri di Milano ds964_nil_wm.zip
quartieri = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/ds964_nil_wm (1)/NIL_WM.dbf')
quartieri
```

```
# installare (se necessario) e importare contextily e visualizzare la mappa di Milano con l'indicazione dei quartieri in nero (solo confini)
# e le piste ciclabili
!pip install contextily
import contextily as ctx 
```

```
ax = quartieri.to_crs(epsg=3857).plot(facecolor='none',edgecolor='k',figsize=(15,10))
bike_ciclabili.to_crs(epsg=3857).plot(ax=ax)
ctx.add_basemap(ax)
```

```
# trovare il quartiere meno esteso
quartieri[quartieri['Shape_Area'] == min(quartieri['Shape_Area'])]
```

```
# trovare il quartiere più esteso
quartieri[quartieri['Shape_Area'] == max(quartieri['Shape_Area'])]
```

```
# visualizzare i due quartieri sulla mappa di Milano con i colori indicati nella figura sotto
ax = quartieri[quartieri['Shape_Area'] == max(quartieri['Shape_Area'])].to_crs(epsg=3857).plot(facecolor='yellow', alpha=0.75, figsize=(15,10))
quartieri[quartieri['Shape_Area'] == min(quartieri['Shape_Area'])].to_crs(epsg=3857).plot(ax=ax, alpha=0.75, facecolor='red')
ctx.add_basemap(ax)
```

```
# salvare in un geodataframe le informazioni sul quartiere di Affori, prendendole dal geodataframe dei quartieri
affori = quartieri.loc[quartieri['NIL'].str.contains('AFFORI')]
affori
```

```
# trovare le piste ciclabili che passano per il quartiere di Affori
```

```
bike_ciclabili[bike_ciclabili.intersects(affori.geometry.squeeze())]
```

```
# visualizzare su una mappa il quartiere di Affori e le piste ciclabili che lo attraversano
# evidenziare bene in rosso le piste ciclabili come nell'esempio sotto
```

```
ax = bike_ciclabili[bike_ciclabili.intersects(affori.geometry.squeeze())].to_crs(epsg=3857).buffer(7).plot(facecolor='r', figsize=(15,10))
affori.to_crs(epsg=3857).plot(ax=ax, facecolor='none', edgecolor='k')
ctx.add_basemap(ax)
```
